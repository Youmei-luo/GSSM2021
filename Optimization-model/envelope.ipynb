####  効率値計算 評価対象企業 ユーグレナ

```python
import pulp
import sys

prob = pulp.LpProblem(name='GSSM', sense=pulp.LpMaximize)
u1 = pulp.LpVariable('u1', lowBound=0.0)
u2 = pulp.LpVariable('u2', lowBound=0.0)
v1 = pulp.LpVariable('v1', lowBound=0.0)
prob += 13317*u1 - 1486*u2  # 目的関数 ユーグレナ
prob += 13317*u1 - 1486*u2 <=  357*v1 , 'ineq1'
prob += 1381997*u1 + 126142*u2 <= 33409*v1 , 'ineq2'
prob += 181287*u1 + 16743*u2 <= 6596*v1 , 'ineq3'
prob += 2192519*u1 + 181205*u2 <= 73275*v1 , 'ineq4'
prob += 920888*u1 - 11660*u2 <= 39035*v1  , 'ineq5'
prob += 114909*u1 + 8016*u2 <= 1277*v1  , 'ineq6'
prob += 279389*u1 + 11986*u2 <= 8768*v1  , 'ineq7'
prob += 9101*u1 - 621*u2 <= 971*v1  , 'ineq8'
prob += 176311*u1 + 4632*u2 <= 4374*v1  , 'ineq9'
prob += 150514*u1 + 19205*u2 <= 3473*v1  , 'ineq10'
prob += 357*v1 <= 1


print(prob) # インスタンスを出力
prob.solve() # 求解
# 結果の表示
print(pulp.LpStatus[prob.status])
print('Optimal Value = ', pulp.value(prob.objective))

for v in prob.variables():
    print(v.name, '=', pulp.value(v))
    if v.name == 'u1':
        u1 = pulp.value(v)
    if v.name == 'u2':
        u2 = pulp.value(v)
    if v.name == 'v1':
        v1 = pulp.value(v)
theta = pulp.value(prob.objective)
result_list = []
result_list.append(['ユーグレナ',theta,u1,u2,v1])

result_list
```
